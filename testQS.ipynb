```python
from numpy.random import randint
```

```python
k = 0
test_cases = []
for i in range(0, 20):
    k= k + 10
    s = '';
    s = str(k) + '\n'
    arr = randint(1, 10**5, k)
    for x in arr:
        s += str(x) + ' '
    s += '\n'
    test_cases.append(s)
```

```python
with open('QS.txt', 'w') as f:
    for test in test_cases:
        f.write(test)
```
