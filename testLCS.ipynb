```python
import numpy as np
import pandas as pd
import string
import random
```

```python
k = 0
test_cases = []
for i in range(0, 20):
    s1 = ''.join(random.choices(string.ascii_uppercase +string.digits, k = i + 5))
    s2 = s1[-5:]
    res = s1 + ' ' + s2 + '\n'
    test_cases.append(res)
```

```python
with open('lcs.txt', 'w') as f:
    for test in test_cases:
        f.write(test)
```
